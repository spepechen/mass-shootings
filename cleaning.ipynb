```python
%matplotlib inline 

import pandas as pd
import matplotlib.pyplot as plt 
from StringIO import StringIO
```

```python
df = pd.read_csv("Copy of Mother Jones' Investigation- US Mass Shootings, 1982-2015 - cleanup.csv")
```

source: 
http://www.motherjones.com/politics/2012/12/mass-shootings-mother-jones-full-data

And I added latest Oregon Community College shooting.

```python
df[:10]
```

```python
df['case'] = df['case'].apply(lambda x: x + ', ')
```

```python
df['case']
```

http://stackoverflow.com/questions/16684346/pandas-group-by-operations-on-a-data-frame

```python
grouped = df.groupby('year')
f_series = grouped.fatalities.sum()
f_series
```

```python
i_series = grouped.injured.sum()
i_series
```

```python
v_series = grouped.total_victims.sum()
v_series
```

http://stackoverflow.com/questions/17841149/pandas-groupby-how-to-get-a-union-of-strings

```python
c_series = grouped['case'].apply(lambda x: x.sum())
c_series 
```

http://stackoverflow.com/questions/18062135/combining-two-series-into-a-dataframe-in-pandas

```python
result = pd.concat([f_series, i_series,v_series, c_series], axis=1)
```

```python
result
```

 http://stackoverflow.com/questions/13682044/pandas-dataframe-remove-unwanted-parts-from-strings-in-a-column

```python
result['case'] = result['case'].map( lambda x: x.rstrip(',')) 
```

```python
result
```
